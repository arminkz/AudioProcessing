```python
import numpy as np
from scipy.fftpack import fft , fftfreq
from scipy.io import wavfile
import matplotlib.pyplot as plt
import IPython.display as ipd

fs, data = wavfile.read('voices/v1.wav')
print(f"sound data shape: {data.shape}")
length = data.shape[0] / fs
print(f"sound length: {length}s")

if data.dtype == 'uint8':
    print('8bit PCM')
if data.dtype == 'int16':
    print('16bit PCM')
if data.dtype == 'int32':
    print('32bit PCM')
if data.dtype == 'float32':
    print('32bit Floating Point')
    
ipd.Audio(data, rate=fs)
```

```python
#get a 5 second sample from sound
data = data[:5*fs]
print(f"sound data shape: {data.shape}")
length = data.shape[0] / fs
print(f"sound length: {length}s")
```

```python
#normalize to [-1,1) range
ndata = data.astype(np.float32)
ndata = (ndata / np.max(np.abs(ndata)))
ndata -= np.mean(ndata)
```

```python
#plot audio
time = np.linspace(0., length, ndata.shape[0])
plt.plot(time, ndata, label="Audio")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
#perform fft
cf = fft(ndata)
d = int(len(cf)/2)-1
cf = cf[:d]

rf = abs(cf)

freqs = fftfreq(ndata.shape[0],1/fs)
freqs = freqs[:d]

idx = np.argsort(freqs)
```

```python
#spectrum graph
plt.plot(freqs[idx],rf[idx],'r')
plt.show()
```

```python
#draw logarithmic graph
plt.xlim( [10, ndata.shape[0]/2] )
plt.xscale('log')
plt.grid(True)
plt.xlabel('Frequency (Hz)')
plt.plot(freqs,rf,'g')
plt.show()
```

```python
maxf = np.where(rf == np.amax(rf))
print(f"max frequency : {freqs[maxf]}")
```
